```python
import requests
from bs4 import BeautifulSoup
import csv
import time
# Базовый URL для отзывов
base_url = "https://www.banki.ru/services/responses/list"

# Заголовки для запроса, чтобы имитировать поведение браузера
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

# Функция для получения отзывов с одной страницы
def get_reviews(page_number):
    url = f"{base_url}?page={page_number}"
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"Ошибка запроса: статус {response.status_code}")
        return []
    classes = []
    soup = BeautifulSoup(response.text, 'html.parser')
    print(str(soup))
    reviews = []
    print(soup.find_all('script')[-1])
    # for div in soup.find_all('div'):
    #     if div.get('class'):
    #         for class_ in div.get('class'):
    #            classes.append(class_) 

    for class_ in list(set(classes)):
        if 'Review' in class_:
            print(class_)    

    print(reviews)
    return reviews

# Основная логика для обхода страниц и сохранения данных
def scrape_all_reviews(start_page=1, end_page=5, start_idx=13999):
    all_reviews = []
    
    for page in range(start_page, end_page + 1):
        print(f"Обработка страницы {page}")
        reviews = get_reviews(page)
        if not reviews:
            print(f"На странице {page} отзывы не найдены или произошла ошибка.")
            continue

        # Фильтрация отзывов по индексу
        filtered_reviews = [r for r in reviews if int(r['idx']) >= start_idx]
        all_reviews.extend(filtered_reviews)
        print(f"Собрано {len(filtered_reviews)} отзывов с страницы {page}")
        
        # Задержка перед следующим запросом
        time.sleep(2)
    
    return all_reviews

# Сохранение данных в CSV файл
def save_to_csv(reviews, filename="reviews.csv"):
    if not reviews:
        print("Нет данных для сохранения.")
        return

    keys = reviews[0].keys()
    with open(filename, 'w', newline='', encoding='utf-8') as output_file:
        dict_writer = csv.DictWriter(output_file, fieldnames=keys)
        dict_writer.writeheader()
        dict_writer.writerows(reviews)

# Пример использования
all_reviews = scrape_all_reviews(start_page=1, end_page=5, start_idx=13999)
save_to_csv(all_reviews)
print(f"Всего собрано {len(all_reviews)} отзывов")
```
